```python
def set_axis_style(ax, labels):
    ax.get_xaxis().set_tick_params(direction='out')
    ax.xaxis.set_ticks_position('bottom')
    ax.set_xticks(np.arange(1, len(labels) + 1))
    ax.set_xticklabels(labels)
    ax.set_xlim(0.25, len(labels) + 0.75)
#     ax.set_xlabel('Sample name')

# data = [sorted(np.random.normal(0, std, 100)) for std in range(1, 4)]
data = [sorted(np.random.normal(0, std, 100)) for std in [1, 0.5, .1]]
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16, 8), sharey=True)
ax.set_title('Formação acadêmica')
ax.set_ylabel('período de formação')


parts = ax.violinplot(data)

for pc in parts['bodies']:
#     pc.set_facecolor('blue')
    pc.set_edgecolor('black')
    pc.set_alpha(.9)

# set style for the axes
labels = ['Graduação\nEconomia', 'Mestrado\nEconomia Aplicada', 'Especialização\nData Science e Big Data']
set_axis_style(ax, labels)
```
